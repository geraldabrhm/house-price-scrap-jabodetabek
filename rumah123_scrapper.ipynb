# Rumah 123 Scrapper

```python
import requests
import bs4
import pandas as pd
```

```python

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36',
    'Accept-Language':'en-US,en;q=0.5',
    'cookie':'locale=en_US;'
}

scraped_data = pd.DataFrame(columns=['title', 'price', 'lt', 'lb'])

def scrape_data(location, page):

    global scraped_data

    _url = f'https://www.rumah123.com/jual/{location}/rumah/?page={page}'

    url = _url
    r = requests.get(url,headers=headers)
    soup = bs4.BeautifulSoup(r.text,'html.parser')

    for item in soup.find('div',{'class':'ui-search-page__content'}).find_all('div',{'class':'card-featured__middle-section'}):
         
        title = ''
        price = ''
        lt = ''
        lb = ''

        if item.find('h2'):
            title = item.find('h2').text

        if item.find('strong'):
            price = item.find('strong').text
        
        tipe = item.find('div',{'class':'card-featured__middle-section__header-badge'}).find('div').text.lower()
        if tipe == 'rumah':
            if item.find('div',{'class':'card-featured__middle-section__attribute'}):
                lb = item.find('div',{'class':'card-featured__middle-section__attribute'}).find_all('div',{'class':'attribute-info'})[-1].text.replace(':','')
                lt = item.find('div',{'class':'card-featured__middle-section__attribute'}).find_all('div',{'class':'attribute-info'})[-2].text.replace(':','')
        
        data = {
            'title': title,
            'price': price,
            'lb' : lb,
            'lt' : lt
        }

        scraped_data = scraped_data.append(data, ignore_index=True)

for i in range(1, 101):
    print(f'Scraping page {i} ...')
    # change the location
    scrape_data('dki-jakarta', i)

scraped_data.to_csv(f'rumah123_dki-jakarta.csv', index=False)
```
